```python
import pandas as pd
```

```python
df_dia_1 = pd.read_csv("../data/dynamic/dia/Dia-categories-url.csv")
df_dia_1.columns = [col.strip() for col in df_dia_1.columns]
df_dia_3 = df_dia_1[['Cat', 'Subcat', 'url', 'page']]
df_dia_3.columns
```

```python
df_dia_3.columns
```

```python
df_dia_2 = pd.read_csv("../data/dynamic/dia/Dia-products.csv")
df_dia_2
```

### Merge and basic statistics (most expensive items)###

```python
df_dia = pd.merge(df_dia_3, df_dia_2, left_index=True, right_on="id")
df_dia.head()
df_dia.to_csv("../data/dynamic/dia/final_table_dia.csv")
```

```python
df_dia.columns
```

```python
price_name = df_dia.sort_values(by="price", ascending=False)[["price", "display_name"]]
cat_price_stat = df_dia.groupby("Subcat")["price"].describe()
cat_price_stat
```

```python
import numpy as np
def get_argmin_cat(x):
    return x[x["price"] == x["price"].min()]
    
cat_groupby = df_dia.groupby("Subcat").apply(get_argmin_cat)[["display_name", "price"]]

cat_groupby
```

```python
import numpy as np

# Find the index of the minimum price within each group
idx_min_price = df_dia.groupby("Subcat")["price"].idxmin()

# Filter the DataFrame to get rows corresponding to the minimum price indices
cat_groupby_min = df_dia.loc[idx_min_price, ["Cat", "Subcat", "display_name", "price", "measure_unit", "price_per_unit"]]

# Reset index if needed
# cat_groupby.reset_index(drop=True, inplace=True)

cat_groupby_min
```

```python
idx_max_price = df_dia.groupby("Subcat")["price"].idxmax()
cat_groupby_max = df_dia.loc[idx_max_price, ["Cat", "Subcat", "display_name", "price", "measure_unit", "price_per_unit"]]
cat_groupby_max
```
